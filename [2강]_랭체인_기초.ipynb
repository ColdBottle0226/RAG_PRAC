# 2. Structured Outputs : 다음 작업을 위해 답변을 특정 형태로 파싱할 때 사용

### Pydantic : 데이터 검증, 직렬화 (타입도 검증함)

```python
#1. 스키마(class) 생성
#2. model에 with_structured_output을 사용해 새로운 모델 객체 생성

#ex) 스키마 생성
from pydantic import BaseModel, Field

class Movie(BaseModel):
  """상세한 영화 정보."""
  title: str = Field(description="영화 제목")
  year: int = Field(description="영화 출시 연도")
  director: str = Field(description="영화 감독")
  rating: float = Field(description="영화 평점")

model_with_structure = model.with_structured_output(Movie)
response = model_with_structure.invoke("영화 인셉션에 대해 설명해주세요")
```

```python
response
```

### JSON output parser

```python
# 영어/숫자/언더스코어/대시만 사용 가능
# pydantic은 파싱 + 데이터 검증까지 지원, JSON Schema는 파싱만 지원
```

```python
import json

json_schema = {
    "title": "Movie",
    "description": "A movie with details",
    "type": "object",
    "properties": {
        "title": {
            "type": "string",
            "description": "The title of the movie"
        },
        "year": {
            "type": "integer",
            "description": "The year the movie was released"
        },
        "director": {
            "type": "string",
            "description": "The director of the movie"
        },
        "rating": {
            "type": "number",
            "description": "The movie's rating out of 10"
        }
    },
    "required": ["title", "year", "director", "rating"]
}
```

```python
response = model.with_structured_output(json_schema).invoke("영화 인셉션에 대해 설명해주세요")
```

```python
response
```

# 3. Memory (메모리 구현)

```python
from langchain_core.messages import AIMessage
## 메시지 구성 요소
#1. System Message : 개발자가 사전에 작성한 입력 (프롬프트)
#2. Human Message : 사용자의 요청(request), 입력
#3. AI Message : Model의 답변(result)

#ex)
[
    SystemMessage("당신은 유능한 로켓 전문가입니다.")
    HumanMessage("안녕하세요. 궁금한 게 있어요!")
    AIMessage("로켓의 기본 원리나 추진 발사 차이 등...")
]
```

### 방법 1 (Human, System, AI 메시지 객체 사용)

```python
from langchain.chat_models import init_chat_model
from langchain.messages import HumanMessage, SystemMessage, AIMessage

model = init_chat_model("gpt-5-nano")

system_msg = SystemMessage("당신은 유능한 로켓 전문가입니다.")
human_msg = HumanMessage("안녕하세요. 궁금한 게 있어요!")

messages = [system_msg, human_msg]

response = model.invoke(messages) # AIMessage 반환

print(response)
```

```python
## 과거 대화 내용을 기억하여 실행하는 방법
```

```python
from langchain.messages import HumanMessage, SystemMessage, AIMessage

human_msg1 = HumanMessage("안녕하세요. 저는 Jay입니다")
ai_msg1 = AIMessage("안녕하세요 Jay! 어떤 도움이 필요하신가요?")
human_msg2 = HumanMessage("제 이름이 뭐라고 했죠?")

messages = [human_msg1, ai_msg1, human_msg2]
response = model.invoke(messages)
print(response)
```

### 방법 2 (딕셔너리 포맷 사용)
